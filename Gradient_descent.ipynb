```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../ML/test_scores.csv')
df.head()
```

```python
df.columns
```

```python
import math
```

```python
def gradient_des(u,v):
    n = len(u)
    m_c, b_c = 0, 0
    iterations = 1000000
    l_rate = 0.0002
    threshold = 1e-20
    cost_threshold = 1e-20
    previous_cost = float('inf')

    
    for i in range(iterations):
        v_p = m_c * u + b_c
        cost = (1/n)*sum([val**2 for val in (v-v_p)])
        m_dv = -(2/n)*sum(u*(v-v_p))
        b_dv = -(2/n)*sum(v-v_p)

        if math.isclose(m_dv, 0, abs_tol = threshold) and math.isclose(b_dv, 0, abs_tol = threshold):
            print(f"Converge at iteration: {i}")
            break

        
        # Stop if cost barely changes
        if abs(cost - previous_cost) < cost_threshold:
            print(f"Converged at iteration {i} by cost change")
            break
        previous_cost = cost
        m_c = m_c - l_rate * m_dv
        b_c = b_c - l_rate * b_dv
        print(f"m: {m_c}, b: {b_c}, cost: {cost}, i: {i}")
    return m_c, b_c
```

```python
u = df['math'].values
v = df['cs'].values
m, b = gradient_des(u, v)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['math']], df.cs)
```

```python
m_learn = reg.coef_
b_learn = reg.intercept_
```

```python
print(f"Using gradient descent: m {m} and b {b}")
print(f"Using sklearn: coef_ {m_learn} and intercept {b_learn}")
```

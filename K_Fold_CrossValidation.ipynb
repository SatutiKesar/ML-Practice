Problem: Use iris flower dataset from sklearn library and use cross_val_score against following models to measure the performance of each.In the end figure out the model with best performance
1. Logistic Regression
2. SVM
3. Decision Tree
4. Random Forest

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(solver = 'lbfgs', max_iter=1000), iris.data, iris.target, cv=3)
```

```python
score_a = cross_val_score(LogisticRegression(solver = 'lbfgs', max_iter=1000), iris.data, iris.target, cv=10)
np.average(score_a)
```

```python
score_b = cross_val_score(SVC(gamma='auto'), iris.data, iris.target, cv=10)
np.average(score_b)
```

```python
from sklearn.tree import DecisionTreeClassifier
score_c = cross_val_score(DecisionTreeClassifier(), iris.data, iris.target, cv=10)
np.average(score_c)
```

```python
score_d = cross_val_score(RandomForestClassifier(n_estimators=20), iris.data, iris.target, cv=10)
np.average(score_d)
```

The best score so far is given by SVM

Problem: Predict canada's per capita income in year 2020 by using canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../ML/canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df['year'], df['per capita income (US$)'], color='maroon', marker='+')
plt.title("Year Vs Price Per Capita (US$)")
plt.show()
```

```python
df.isnull().sum()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
reg.intercept_
```

```python
reg.coef_
```
